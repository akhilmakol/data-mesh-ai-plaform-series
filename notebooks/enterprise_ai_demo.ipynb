
# 🚀 Enterprise AI Demo — Data Mesh AI Platform

This notebook demonstrates an **end-to-end AI workflow** based on the repository architecture.

Steps demonstrated:

1. Load domain datasets (customers, loans, payments)
2. Perform basic feature engineering
3. Train a simple ML risk model
4. Run an AI decision agent
5. Produce a loan approval decision

This simulates a **Data Mesh × Enterprise AI platform workflow**.


```

import pandas as pd
from sklearn.linear_model import LogisticRegression
```

## Step 1 — Load Domain Data Products

```

customers = pd.read_csv("../datasets/customers.csv")
loans = pd.read_csv("../datasets/loans.csv")
payments = pd.read_csv("../datasets/payments.csv")

print("Customers")
display(customers)

print("Loans")
display(loans)

print("Payments")
display(payments)
```

## Step 2 — Feature Engineering

```

features = customers.copy()

features["income_to_credit_ratio"] = features["income"] / features["credit_score"]

display(features)
```

## Step 3 — Prepare Training Data

```

# Mock target variable for demo
features["loan_default"] = [0,1,0]

X = features[["income","credit_score","age"]]
y = features["loan_default"]

X, y
```

## Step 4 — Train AI Model

```

model = LogisticRegression()
model.fit(X, y)

print("Model trained successfully")
```

## Step 5 — Run AI Inference

```

sample_customer = [[70000,710,35]]

prediction = model.predict(sample_customer)
probability = model.predict_proba(sample_customer)

prediction, probability
```

## Step 6 — Agentic AI Decision Layer

```

def loan_agent(prob):

    if prob > 0.6:
        return "Reject Loan"
    
    return "Approve Loan"

decision = loan_agent(probability[0][1])

print("AI Agent Decision:", decision)
```


## 🎯 Result

We simulated an **enterprise AI workflow**:

Domain Data → Feature Engineering → ML Model → AI Agent Decision

This demonstrates how **Data Mesh architectures can power enterprise AI platforms**.
